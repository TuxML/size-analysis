```python
import pandas as pd
```

```python
df = pd.read_csv("results/results.csv")
df.head()
```

```python
#get the random_id for the best mean
random_id = df.sort_values("mean").iloc[0]["random_id"]
random_id
```

```python
df2 = df[df["random_id"] == random_id].sort_values("train_size")
df2
```

```python
df2.plot( x="train_size", y="mean", style="o-", ylim=(0,100), figsize=(12,12))
```
