```python
import numpy as np
import pandas as pd

list_csv=[
    "config_bdd15000-20000.csv",
    "config_bdd20000-30000.csv",
    "config_bdd30000-40000.csv",
    "config_bdd40000-60000.csv",
    "config_bdd60000-90000.csv",
    "config_bdd90000-100000.csv",
    "config_bdd100000-1265000.csv",
]

ftuniques = []
freq_ymn_features = []
non_tristate_options = []
for csv in list_csv:

    df = pd.read_csv("../tuxml-datasets/"+csv)
    
    
    basic_head = ["cid", "time", "date"] # "compile"
    size_methods = ["vmlinux", "GZIP-bzImage", "GZIP-vmlinux", "GZIP", "BZIP2-bzImage", 
                  "BZIP2-vmlinux", "BZIP2", "LZMA-bzImage", "LZMA-vmlinux", "LZMA", "XZ-bzImage", "XZ-vmlinux", "XZ", 
                  "LZO-bzImage", "LZO-vmlinux", "LZO", "LZ4-bzImage", "LZ4-vmlinux", "LZ4"]
    compilation_status_column_name = 'kernel_size'
    
    
    #Filtering out X86_32 configuration
    df.query("X86_64 == 'y'", inplace=True)
    
    
    #tri state step
    tri_state_values = ['y', 'n', 'm']


    for col in df:
        ft = df[col]    
        # eg always "y"
        if len(ft.unique()) == 1:
            ftuniques.append(col)
        # only tri-state values (y, n, m) (possible TODO: handle numerical/string options)    
        elif all(x in tri_state_values for x in ft.unique()):     #len(ft.unique()) == 3: 
            freq = ft.value_counts(normalize=True)
            freqy = 0
            freqn = 0
            freqm = 0
            if ('y' in freq.index):
                freqy = freq['y']
            if ('n' in freq.index):
                freqn = freq['n']
            if ('m' in freq.index):
                freqm = freq['m']
            freq_ymn_features.append((col, freqy, freqm, freqn))
        else:
            if not (col in size_methods): 
                non_tristate_options.append(col)
```

```python
non_tristate_options = list(set(non_tristate_options) - set(basic_head))
```

```python
possible_values = {}

for csv in list_csv:

    df = pd.read_csv("../tuxml-datasets/"+csv)
    
    #Filtering out X86_32 configuration
    df.query("X86_64 == 'y'", inplace=True)
    df.drop(columns=basic_head, inplace=True)
    
    for i in list(set(non_tristate_options)):
        if not i in possible_values:
            possible_values[i] = []
        try:
            possible_values[i] = list(set(possible_values[i] + list(df[i].unique())))
        except Exception as e:
            print(e)
            print(i)
            print(possible_values[i])
            print(list(df[i].unique()))
            break
```

```python
for i in possible_values:
    if np.NaN in possible_values[i]:
        possible_values[i] = [None if type(j)==np.NaN else j for j in possible_values[i]]
    if type(possible_values[i][0]) == np.int64:
        possible_values[i] = [int(j) for j in possible_values[i]]
```

```python
import json
with open("non_tristate_options.json","w") as f:
    json.dump(possible_values, f)
```
