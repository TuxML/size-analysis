```python
import tuxml
import pandas as pd

df = tuxml.load_dataset()
df_options = pd.read_csv("dataOptionsText/comparison/feature_importanceRF/optionsInBoth.csv")
df_options.sort_values(by="Ranking")

freq = []
for o in df_options['option']:
    freq.append(df[o].value_counts(normalize=True)[0])
df_options["freq"] = freq
df_options
```
