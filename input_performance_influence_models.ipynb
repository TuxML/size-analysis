```python
import tuxml
import pandas as pd 

df = tuxml.load_dataset()
```

```python
df.sort_values("vmlinux")[:5]
```

```python
size_methods = ["vmlinux", "GZIP-bzImage", "GZIP-vmlinux", "GZIP", "BZIP2-bzImage", 
              "BZIP2-vmlinux", "BZIP2", "LZMA-bzImage", "LZMA-vmlinux", "LZMA", "XZ-bzImage", "XZ-vmlinux", "XZ", 
              "LZO-bzImage", "LZO-vmlinux", "LZO", "LZ4-bzImage", "LZ4-vmlinux", "LZ4"]
additional_ft = ["nbyes", "nbno", "nbmodule", "nbyesmodule"]
```

```python
import xml.etree.cElementTree as ET

df = df.drop(columns=["cid"]).drop(columns=size_methods)[:500]
#columns = list(df.drop(columns=["cid"]).drop(columns=size_methods).columns.values)
columns = list(df.columns.values)
root = ET.Element("results")

for row in df.iterrows():
    config=""
    
    for i in columns:
        if (df[i][row[0]]==1):
            config=config+i+","
    
    doc = ET.SubElement(root, "row")

    ET.SubElement(doc, "data", columname="Configuration").text = config
    ET.SubElement(doc, "data", columname="Performance").text = str(df["vmlinux"][row[0]])

tree = ET.ElementTree(root)
tree.write("measurements.xml")
```
